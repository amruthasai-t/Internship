```python
trending=driver.find_element(By.XPATH,"/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[3]/div/div[3]/ul/li[2]/a")
```

```python
trending.click()
```

```python
rtitles=[]
rnametags=driver.find_elements(By.XPATH,'//span[@class="text-normal"]')
for i in rnametags:
    rtitles.append(i.text.strip(' /'))
rdescs=[]
rcount=[]
languages=[]
rcountags=driver.find_elements(By.XPATH,'//a[2][@class="Link Link--muted d-inline-block mr-3"]')
for i in rcountags:
    rcount.append(i.text.strip())
rdesctags = driver.find_elements(By.XPATH, '//article[@class="Box-row"]')
for i in rdesctags:
    try:
        if i.find_element(By.XPATH,'//p[@class="col-9 text-gray my-1 pr-4"]'):
            r=i.find_element(By.XPATH,'//p[@class="col-9 text-gray my-1 pr-4"]')
            rdescs.append(r.text.strip())
    except NoSuchElementException:
        rdescs.append('-')
langtags = driver.find_elements(By.XPATH, '//div[@class="f6 color-fg-muted mt-2"]')
for i in langtags:
    try:
        if i.find_element(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]/span[2]'):
            l=i.find_element(By.XPATH,'//span[@class="d-inline-block ml-0 mr-3"]/span[2]')
            languages.append(l.text.strip())
    except NoSuchElementException:
        languages.append('-')
```

```python
ranks = [num for num in lwrprwb if num]

lwr=ranks[::3]
pr=ranks[1::3]
wb=ranks[2::3]
```

```python
df=pd.DataFrame({'Name':names,'Artist':artists,'Last Week Rank':lwr,'Peak Rank':pr,'Weeks on board':wb})
df
```

6. Scrape the details of Highest selling novels.
A) Book name
B) Author name
C) Volumes sold
D) Publisher
E) Genre
 Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare


```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")
```

```python
table = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[2]/div/div[2]/div/table')
ranks=[]
book_names=[]
author_names=[]
volumes_sold=[]
publishers=[]
genres=[]
rows = table.find_elements(By.TAG_NAME, "tr")
for row in rows[2:]:
    columns = row.find_elements(By.TAG_NAME, "td")
    ranks.append(columns[0].text)
    book_names.append(columns[1].text)
    author_names.append(columns[2].text)
    volumes_sold.append(columns[3].text)
    publishers.append(columns[4].text)
    genres.append(columns[5].text)
```

```python
for i in range(len(book_names)):
    print("Book Name:",book_names[i])
    print("Author Name:",author_names[i])
    print("Volumes Sold:",volumes_sold[i])
    print("Publisher:",publishers[i])
    print("Genre:",genres[i])
    print()
```

```python
driver.quit()
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
datasets=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
datasets.click()
```

```python
dnametags=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
dnames=[]
for i in dnametags:
    dnames.append(i.text.strip())
dtypes=[]
dtypetags=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
for i in dtypetags:
    dtypes.append(i.text.strip())
tasks=[]
tasktags=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1]/span')
for i in tasktags:
    tasks.append(i.text.strip())
instances=[]
instags=driver.find_elements(By.XPATH,'//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
for i in instags:
    instances.append(i.text.strip(""))
attributes=[]
attritags=driver.find_elements(By.XPATH,'/html/body/div[1]/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[1]/div[2]/svg')
for i in attritags:
    attributes.append(i.text.strip(""))
```

```python
df=pd.DataFrame({'Name':dnames,'Task':tasks,'No. of instances':instances,'No. of attributes':attributes,'Datatype':dtypes})
df
```
