```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

1) Write a python program to display all the header tags from wikipedia.org.


```python
def get_header_tags(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    titles=[]
    tags=['h1','h2','h3','h4','h5','h6']
    for i in soup.find_all(tags):
        titles.append(i.text)
    return titles
url='https://en.wikipedia.org/wiki/Main_Page'
get_header_tags(url)
```

2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year
of release) and make data frame.

```python
def get_100_movies(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content, 'html.parser')
    titles=[]
    years=[]
    ratings=[]
    for i in soup.find_all('h3', class_='lister-item-header'):
        titles.append(i.find('a').text.strip())
    for i in soup.find_all('span', class_='lister-item-year'):
        year_text=i.text.strip()
        year = re.search(r'\d{4}', year_text).group()
        years.append(year)
    ratings=[]
    for i in soup.find_all('div', class_='ipl-rating-star small'):
        rating = i.find('span', class_="ipl-rating-star__rating").text.strip()
        ratings.append(rating)
    df=pd.DataFrame({'Name':titles,'Rating':ratings,'Year of Release':years})
    return df

imdb_url = 'https://www.imdb.com/list/ls004139640/'
get_100_movies(imdb_url)
```

3) Write a python program to scrape mentioned details from dineout.co.in: i) Restaurant name
ii) Cuisine iii) Location iv) Ratings v) Image URL.



```python
def get_restaurants(url):
    page=requests.get(url)
    soup=BeautifulSoup(page.content, 'html.parser')
    names=[]
    cuisine=[]
    locs=[]
    ratings=[]
    img_urls=[]
    for i in soup.find_all('a', class_='restnt-name ellipsis'):
        names.append(i.text)
    for i in soup.find_all('div',class_='img-wrap'):
        ratings.append(i.find('div',class_='restnt-rating').text)
    for i in soup.find_all('img', class_="no-img"):
        img_urls.append(i['data-src'])
    for i in soup.find_all('div', class_='restnt-loc ellipsis'):
        locs.append(i.text.strip())
    for i in soup.find_all('span', class_='double-line-ellipsis'):
        a=[]
        for j in i.find_all('a'):
            a.append(j.text)
        cuisine.append(a)
    df=pd.DataFrame({'Restaurant Name':names,'Cuisines':cuisine,'Location':locs,'Rating':ratings,'Image urls':img_urls})
    return df
url="https://www.dineout.co.in/hyderabad-restaurants/trending"
get_restaurants(url)
```
